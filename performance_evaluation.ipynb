```python
df = pd.DataFrame(
    {'dataset': dataset,
     'transformation': transformation,
     'classifier': classifier,
     'accuracy': accuracy,
     'mcc': mcc
    })

df
```

```python
# Accuracy and Mathews' Correlation Coefficient
fig, axes = plt.subplots(2, 3, figsize=(14, 10), sharey=False)

sns.set(font_scale = 1.5)

box1 = sns.boxplot(ax = axes[0,0], data=df, x="dataset", y="accuracy", color = "lightgreen")
box2 = sns.boxplot(ax = axes[0,1], data=df, x="transformation", y="accuracy", color = "skyblue", whis = 3)
box3 = sns.boxplot(ax = axes[0,2], data=df, x="classifier", y="accuracy", color = "coral")
box4 = sns.boxplot(ax = axes[1,0], data=df, x="dataset", y="mcc", color = "lightgreen")
box5 = sns.boxplot(ax = axes[1,1], data=df, x="transformation", y="mcc", color = "skyblue")
box6 = sns.boxplot(ax = axes[1,2], data=df, x="classifier", y="mcc", color = "coral")

# box1.set_xticklabels(['R \n (csv)','Python \n (csv)','JAVA \n (csv)', 'DTP \n (csv)', 'DTP \n (nodes)'])
box2.set_xticklabels(['original','normalized \n (original)','eigen \n (k largest)','eigen \n (2k largest)'])
# box3.set_xticklabels(['Dataset \n 1','Dataset \n 2','Dataset \n 3'])
# box4.set_xticklabels(['R \n (csv)','Python \n (csv)','JAVA \n (csv)', 'DTP \n (csv)', 'DTP \n (nodes)'])
box5.set_xticklabels(['original','normalized \n (original)','eigen \n (k largest)','eigen \n (2k largest)'])
# box6.set_xticklabels(['Dataset \n 1','Dataset \n 2','Dataset \n 3'])

box1.set(xlabel="A")
box2.set(xlabel="B")
box3.set(xlabel="C")
box4.set(xlabel="D")
box5.set(xlabel="E")
box6.set(xlabel="F")

box1.set_ylabel('Accuracy in percentage (%)', fontsize = 18)
box2.set(ylabel=None)
box3.set(ylabel=None)
box4.set_ylabel('Matthews Correlation Coefficient', fontsize = 18)
box5.set(ylabel=None)
box6.set(ylabel=None)

box1.set_xticklabels(box1.get_xticklabels(), size = 14)
box2.set_xticklabels(box2.get_xticklabels(), size = 14, rotation = 45)
box3.set_xticklabels(box3.get_xticklabels(), size = 14)
box4.set_xticklabels(box1.get_xticklabels(), size = 14)
box5.set_xticklabels(box2.get_xticklabels(), size = 14, rotation = 45)
box6.set_xticklabels(box3.get_xticklabels(), size = 14)

# box1.set(ylim=(61, 93))
# box2.set(ylim=(61, 93))
# box3.set(ylim=(61, 93))

fig.subplots_adjust(wspace=0.2, hspace=0.6)
```
