```python
df_train, df_test = train_test_split(df_hf, test_size = 0.25)
X_train, y_train = df_train.iloc[:, :-1], df_train.iloc[:, [-1]]
X_test, y_test = df_test.iloc[:, :-1], df_test.iloc[:, [-1]]
```

```python
def centroidLearningTrain(df_train, class_label):
    df_centroids = df_train.groupby([class_label]).mean()
    return df_centroids
    
```

```python
def centroidLearningTest(features, labels, class_label, df_centroids):
    
    cluster1 = []
    cluster2 = []
    df_output1 = pd.DataFrame()
    df_output2 = pd.DataFrame()
    
    X_centroids, y_centroids = df_centroids, df_centroids.index
    
    
    for label in labels[class_label].unique():
        
        for feature in features:
            dist1 = np.linalg.norm(feature - df_centroids.iloc[0])
            dist2 = np.linalg.norm(feature - df_centroids.iloc[1])
            
            if dist2 > dist1:
                df_output1.append(dist1)
                cluster1.append(dist1)
            else:
                df_output2.append(dist2)
                cluster2.append(dist2)
                
    return cluster1, cluster2
```
